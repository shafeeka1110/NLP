```python
import pandas as pd
import numpy as np
import string
from sklearn.datasets import fetch_20newsgroups
```

```python
categories=['sci.space']
df=fetch_20newsgroups(subset='train',categories=categories)
```

```python
df=pd.DataFrame({'text': df.data})
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
```

```python
def remove_stop(x):
    return " ".join([word for word in x.split() if word.lower() not in stop_words])
df['cleaned_text']=df['text'].apply(lambda x: remove_stop(x))
```

```python
df['cleaned_text'].head()
```

```python
df=' '.join(df['cleaned_text'])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud=WordCloud(width=800,height=400,background_color='orange').generate(df)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```
