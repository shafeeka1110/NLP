```python
import spacy
from sklearn.datasets import fetch_20newsgroups
nlp = spacy.load("en_core_web_sm")
```

```python
newsgroups_data = fetch_20newsgroups(subset='train')  # Use 'train' or 'test'
```

```python
text = newsgroups_data.data[0]
```

```python
doc = nlp(text)
```

```python
print("Tokens:")
for token in doc:
    print(token.text)
```

```python
print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
print("\nPart-of-Speech Tagging:")
for token in doc:
    print(f'{token.text}: {token.pos_}')
```
